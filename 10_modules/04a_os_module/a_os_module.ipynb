````python
# assignment - like a anti-virus scan, scan all the executables in your operating system. Based on OS, executable need to be taken
````

````python
import os
import platform
import fnmatch

# Function to detect OS type
def get_os_type():
    return platform.system().lower()

# Function to scan executables based on OS
def scan_executables():
    os_type = get_os_type()
    
    # Define the patterns for executables based on the OS
    if os_type == 'windows':
        # On Windows, we scan for files with .exe, .bat, .cmd, .msi extensions
        patterns = ['*.exe', '*.bat', '*.cmd', '*.msi']
        path_to_scan = 'C:\\'  # You can modify this to specific directories
    elif os_type == 'linux' or os_type == 'darwin':  # Linux and macOS
        # On Linux/macOS, executables typically don't have extensions, but we can check for executable files
        patterns = ['*']
        path_to_scan = '/'  # This will scan the entire file system
    else:
        print(f"Unsupported OS: {os_type}")
        return []

    # List to hold the found executables
    executables_found = []

    # Walk through the file system
    for root, dirs, files in os.walk(path_to_scan):
        for pattern in patterns:
            for filename in fnmatch.filter(files, pattern):
                # Full path to the executable file
                file_path = os.path.join(root, filename)
                
                # For Unix-like systems (Linux/macOS), check if the file is executable
                if os_type != 'windows':
                    if os.access(file_path, os.X_OK):
                        executables_found.append(file_path)
                else:
                    executables_found.append(file_path)
    
    return executables_found

# Scan for executables
executables = scan_executables()

# Print the found executables
print(f"Found {len(executables)} executables:")
for exec_file in executables:
    print(exec_file)
````

```
os.pipe(): This function creates a pair of connected file descriptors, which can be used for interprocess communication. This function is not available on Windows.

os.chroot(): This function changes the root directory of the current process to the specified directory. This function is not available on Windows.

os.fchown(): This function changes the owner and group of a file, given the file descriptor. This function is not available on Windows.

os.setresuid(): This function sets the real, effective, and saved user IDs of the current process. This function is not available on Windows.

os.setresgid(): This function sets the real, effective, and saved group IDs of the current process. This function is not available on Windows.
```
